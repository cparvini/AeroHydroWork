```python
import math
import numpy
from matplotlib import pyplot
```

```python
def get_velocity_doublet(strength, xd, yd, X, Y):
    """Returns the velocity field generated by a doublet.
    
    Arguments
    ---------
    strength -- strength of the doublet.
    xd, yd -- coordinates of the doublet.
    X, Y -- mesh grid.
    """
    u = - strength/(2*math.pi)*((X-xd)**2-(Y-yd)**2)/((X-xd)**2+(Y-yd)**2)**2
    v = - strength/(2*math.pi)*2*(X-xd)*(Y-yd)/((X-xd)**2+(Y-yd)**2)**2
    
    return u, v
```

```python
def get_stream_function_doublet(strength, xd, yd, X, Y):
    """Returns the stream-function generated by a doublet.
    
    Arguments
    ---------
    strength -- strength of the doublet.
    xd, yd -- coordinates of the doublet.
    X, Y -- mesh grid.
    """
    psi = - strength/(2*math.pi)*(Y-yd)/((X-xd)**2+(Y-yd)**2)
    
    return psi
```

```python
def get_freestream_info(u_inf, Y, N):
    """Returns the stream-function generated by a freestream.
    
    Arguments
    ---------
    u_inf -- freestream velocity.
    Y -- mesh grid of y values.
    N -- number of discrete points in each direction on our grid.
    """
    u = u_inf * numpy.ones((N, N), dtype=float)
    v = numpy.zeros((N, N), dtype=float)
    psi = u_inf * Y
    
    return u, v, psi
```

```python
def doublet_cylindrical_info(strength, R, Theta):
    """Returns the stream-function & velocities generated by a doublet in
    cylindrical coordinates.
    
    Arguments
    ---------
    strength -- doublet strength.
    R, Theta -- mesh grid of r-theta values.
    """
    
    vr = - (strength*numpy.cos(Theta)/(2*math.pi*(R**2)))
    vtheta = - (strength*numpy.sin(Theta)/(2*math.pi*(R**2)))
    
    psi = - strength*numpy.sin(Theta)/(2*math.pi*R)
    
    return vr, vtheta, psi
```

```python
def freestream_cylindrical_info(u_inf, R, Theta):
    """Returns the stream-function & velocities generated by a freestream in
    cylindrical coordinates.
    
    Arguments
    ---------
    u_inf -- freestream velocity.
    R, Theta -- mesh grid of r-theta values.
    """
    
    vr = u_inf*numpy.cos(Theta)
    vtheta = - u_inf*numpy.sin(Theta)
    
    psi = u_inf*R*numpy.sin(Theta)
    
    return vr, vtheta, psi
```

```python
# initial setup
N = 100
x_start, x_end = -2.0, 2.0
y_start, y_end = -1.0, 1.0
x = numpy.linspace(x_start, x_end, N)
y = numpy.linspace(y_start, y_end, N)
X, Y = numpy.meshgrid(x,y)

kappa = 1.0
u_inf = 1.0
x_doublet, y_doublet = 0.0, 0.0
```

```python
# the math
u_freestream, v_freestream, psi_freestream = get_freestream_info(u_inf, Y, N)
u_doublet, v_doublet = get_velocity_doublet(kappa, x_doublet, y_doublet, X, Y)
psi_doublet = get_stream_function_doublet(kappa, x_doublet, y_doublet, X, Y)
```

```python
# the doublet plotting
%matplotlib inline

size = 10
pyplot.figure(figsize=(size, (y_end-y_start)/(x_end-x_start)*size))
pyplot.xlabel('x', fontsize=16)
pyplot.ylabel('y', fontsize=16)
pyplot.xlim(x_start, x_end)
pyplot.ylim(y_start, y_end)
pyplot.streamplot(X, Y, u_doublet, v_doublet, 
               density=2, linewidth=1, arrowsize=1, arrowstyle='->')
pyplot.scatter(x_doublet, y_doublet, color='#CD2305', s=80, marker='o');
```

```python
# plot the doublet in a freestream
# first, superposition:
u = u_freestream + u_doublet
v = v_freestream + v_doublet
psi = psi_freestream + psi_doublet

# plots the combination
size = 10
pyplot.figure(figsize=(size, (y_end-y_start)/(x_end-x_start)*size))
pyplot.xlabel('x', fontsize=16)
pyplot.ylabel('y', fontsize=16)
pyplot.xlim(x_start, x_end)
pyplot.ylim(y_start, y_end)
pyplot.streamplot(X, Y, u, v, density=2, linewidth=1, arrowsize=1, arrowstyle='->')
pyplot.contour(X, Y, psi, levels=[0.], colors='#CD2305', linewidths=2, linestyles='solid')
pyplot.scatter(x_doublet, y_doublet, color='#CD2305', s=80, marker='o')

# calculates the stagnation points
x_stagn1, y_stagn1 = +math.sqrt(kappa/(2*math.pi*u_inf)), 0
x_stagn2, y_stagn2 = -math.sqrt(kappa/(2*math.pi*u_inf)), 0

# find the radius of the cylinder
r = numpy.sqrt((x_stagn1 - x_doublet)**2 + (y_stagn1 - y_doublet)**2)
print('The radius of the cylinder created is ' + repr(r))

# adds the stagnation points to the figure
pyplot.scatter([x_stagn1, x_stagn2], [y_stagn1, y_stagn2], color='g', s=80, marker='o');
```

```python
# Convert the locations to r-theta
R = numpy.sqrt(X**2 + Y**2)
Theta = numpy.arctan(X/Y)

vr_freestream, vtheta_freestream, psi_freestream = \
        freestream_cylindrical_info(u_inf, R, Theta)
vr_doublet, vtheta_doublet, psi_doublet = \
        doublet_cylindrical_info(kappa, R, Theta)

# superposition:
vr = numpy.round(vr_freestream + vr_doublet,2)
vtheta = numpy.round(vtheta_freestream + vtheta_doublet,2)
psi = psi_freestream + psi_doublet

# plots the combination
size = 10
pyplot.figure(figsize=(size, (y_end-y_start)/(x_end-x_start)*size))
pyplot.xlabel('x', fontsize=16)
pyplot.ylabel('y', fontsize=16)
pyplot.xlim(x_start, x_end)
pyplot.ylim(y_start, y_end)
pyplot.streamplot(X, Y, vr, vtheta, density=2, linewidth=1, arrowsize=1, arrowstyle='->')
pyplot.contour(X, Y, psi, levels=[0.], colors='#CD2305', linewidths=2, linestyles='solid')
pyplot.scatter(x_doublet, y_doublet, color='#CD2305', s=80, marker='o')

# speed of the flow at the surface
ids = numpy.ravel_multi_index(numpy.where(numpy.round(R,2) == numpy.round(r,2)), X.shape)

# check if we grabbed the right coordinates
#pyplot.scatter(X.flat[ids], Y.flat[ids], color='#CD2305', s=80, marker='o')

vr_surface = vr.flat[ids]
vtheta_surface = vtheta.flat[ids]
print('The velocities at the surface are as follows:')
print('V_r: \n' + repr(vr_surface) + '\nV_theta:\n' + repr(vtheta_surface))
print('\nObviously, the no slip condition is not working at the surface!')
```

```python
# computes the pressure coefficient field
cp = 1.0 - (u**2+v**2)/u_inf**2

# plots the pressure coefficient field
size = 10
pyplot.figure(figsize=(1.1*size, (y_end-y_start)/(x_end-x_start)*size))
pyplot.xlabel('x', fontsize=16)
pyplot.ylabel('y', fontsize=16)
pyplot.xlim(x_start, x_end)
pyplot.ylim(y_start, y_end)
contf = pyplot.contourf(X, Y, cp, levels=numpy.linspace(-2.0, 1.0, 100), extend='both')
cbar = pyplot.colorbar(contf)
cbar.set_label('$C_p$', fontsize=16)
cbar.set_ticks([-2.0, -1.0, 0.0, 1.0])
pyplot.scatter(x_doublet, y_doublet, color='#CD2305', s=80, marker='o')
pyplot.contour(X,Y,psi, levels=[0.], colors='#CD2305', linewidths=2, linestyles='solid')
pyplot.scatter([x_stagn1, x_stagn2], [y_stagn1, y_stagn2], color='g', s=80, marker='o');
```

```python
# compute the pressure coefficient for our radial values
cp_surface_actual = 1.0 - (u.flat[ids]**2 + v.flat[ids]**2)/u_inf**2
cp_eqn = 1.0 - 4 * numpy.sin(Theta.flat[ids])**2

print('The actual field values for C_p are:\n' + repr(cp_surface_actual))
print('While the values calculated by the given equation are:\n' + repr(cp_eqn))

print(repr(cp_surface_actual/cp_eqn))
CP = numpy.zeros((N,N))
CP.flat[ids] = cp_surface_actual
ThetaPlot = numpy.degrees(Theta)
# plot
size = 10
pyplot.figure(figsize=(1.1*size, (y_end-y_start)/(x_end-x_start)*size))
pyplot.xlabel('C_p', fontsize=16)
pyplot.ylabel('Theta (degrees)', fontsize=16)
pyplot.xlim(numpy.min(cp_surface_actual), numpy.max(cp_surface_actual))
pyplot.ylim(numpy.min(ThetaPlot.flat[ids]), numpy.max(ThetaPlot.flat[ids]))
pyplot.scatter(cp_surface_actual,ThetaPlot.flat[ids])
```
