```python
import numpy
import cmath
from matplotlib import pyplot
import aero_functions as aero
%matplotlib inline
```

```python
def J_transform(zeta, c):
    """
    Map the complex coordinates into another plane.
    
    Arguments
    ---------
    zeta -- complex; coordinates in the complex plane.
    c -- float; constant value
    
    Returns
    ---------
    z -- complex; coordinates in the new plane
    
    """
    z = numpy.empty_like(zeta)
    z = zeta + ((c**2)/zeta) 
    
    return z

def J_velocity_transform(zeta, zin, c):
    """
    Map the u and v velocities into another plane.
    
    Arguments
    ---------
    zeta -- complex; velocity in the complex plane.
    c -- float; constant value
    
    Returns
    ---------
    z -- complex; velocities in the new plane
    
    """
    z = numpy.empty_like(zeta)
    z = (zeta) / (1-(c/zin)**2) 
    
    return z

def create_rtheta_grid(Nr, Ntheta, r_start, r_end, theta_start, theta_end):
    """
    Create a polar-coordinate grid.
    
    Arguments
    ---------
    Nr, Ntheta -- int; number of points in the r and theta directions.
    theta_start, 
    theta_end,
    r_start, r_end --  float; beginning and end points for the grid.
    
    Returns
    ---------
    R, Theta -- 2D matrix; meshgridded r and theta values.
    
    """
    r = numpy.linspace(r_start, r_end, Nr)
    theta = numpy.linspace(theta_start, theta_end, Ntheta)
    R, Theta = numpy.meshgrid(r, theta)
    
    return R, Theta
```

```python
a = numpy.complex(2.75, 3.69) * numpy.complex(8.55,-6.13)
b = 1.4 * numpy.exp(5.32*1j)
c = numpy.complex(7.51, -9.15) / numpy.complex(4.43,9.64)

print a, b, c
```

```python
# Initial Conditions
N = 100
Nr = 100
Ntheta = 145
c, r = 1., 1.15
u_inf = 1.0
x0, y0 = -0.15, 0.
x_start, x_end, y_start, y_end = -5.0, 5.0, -5.0, 5.0
theta_start, theta_end, r_start, r_end = 0, 2*numpy.pi, r, 5.0 
X, Y = aero.create_grid(N, x_start, x_end, y_start, y_end)
R, Theta = create_rtheta_grid(Nr, Ntheta, r_start, r_end, theta_start, theta_end)
XRT, YRT = R*numpy.cos(Theta)+x0, R*numpy.sin(Theta)+y0

xcircle = numpy.linspace(-r,r,N) + x0
y_up = numpy.sqrt(r**2. - (xcircle-x0)**2.) + y0
y_down = -numpy.sqrt(r**2. - (xcircle-x0)**2.) + y0

# Build the circle
xcomb = numpy.concatenate([xcircle, xcircle]) 
ycomb = numpy.concatenate([y_up, y_down])
coords = numpy.vstack((xcomb, ycomb))
```

```python
# Part 2
zeta = numpy.empty_like(coords, dtype=complex)
zeta.real = coords[0,:]
zeta.imag = coords[1,:]
z = J_transform(zeta, c)
x = z.real
y = z.imag

pyplot.figure(figsize=(12, 12))
pyplot.xlabel('x', fontsize=16)
pyplot.ylabel('y', fontsize=16)
pyplot.xlim(x_start/2, x_end/2)
pyplot.ylim(y_start/2, y_end/2)
pyplot.grid(which='both', axis='both')
pyplot.scatter(coords[0,:], coords[1,:], color='r')
pyplot.scatter(x, y)
```

```python
# Part 3
zeta3 = numpy.empty_like(XRT, dtype=complex)
zeta3.real = XRT
zeta3.imag = YRT
z3 = J_transform(zeta3, c)
xz3, yz3 = z3.real, z3.imag

fig, (ax1, ax2) = pyplot.subplots(1, 2, figsize=(16, 8))
pyplot.xlabel('x', fontsize=16)
pyplot.ylabel('y', fontsize=16)
pyplot.xlim(1.25*x_start, 1.25*x_end)
pyplot.ylim(1.25*y_start, 1.25*y_end)
ax1.scatter(XRT, YRT)
ax2.scatter(xz3, yz3)
```

```python
# Part 4
strength_doublet = 2.*numpy.pi*u_inf*r**2.
vortex_strength = 1.0

# doublet velocities
u_doublet, v_doublet, psi_doublet = aero.get_doublet_info(strength_doublet, x0, y0, XRT, YRT)

# freestream
u_fs, v_fs, psi_fs = aero.get_freestream_info(u_inf, 0.0, XRT, YRT, Ntheta, Nr)

# combine
u = (u_fs + u_doublet)
v = (v_fs + v_doublet)
psi = psi_fs + psi_doublet
cp1 = 1.0 - (u**2 + v**2)/u_inf**2

uc = numpy.empty_like(XRT, dtype=complex)
uc = u - 1j * v
dEdz = 1 - (c/(XRT + 1j * YRT))**2
wE = uc / dEdz
uw = wE.real
vw = -wE.imag
cp = 1.0 - (uw**2 + vw**2)/u_inf**2

# Plotting
# Streamlines
fig, (ax1, ax2) = pyplot.subplots(1, 2, figsize=(16, 8), sharex=True, sharey=True)
pyplot.suptitle('Streamlines for Part 4', size=24)
pyplot.xlabel('x', fontsize=16)
pyplot.ylabel('y', fontsize=16)
pyplot.xlim(1.25*x_start, 1.25*x_end)
pyplot.ylim(1.25*y_start, 1.25*y_end)
ax2.contour(xz3, yz3, psi, colors='k', linestyles='solid', levels=numpy.linspace(numpy.min(psi),numpy.max(psi), 100))
ax1.contour(XRT, YRT, psi, colors='k', linestyles='solid', levels=numpy.linspace(numpy.min(psi),numpy.max(psi), 100))


# Velocity Vectors
fig, (ax1, ax2) = pyplot.subplots(1, 2, figsize=(16, 8), sharex=True, sharey=True)
pyplot.suptitle('Velocity Vectors for Part 4', size=24)
pyplot.xlabel('x', fontsize=16)
pyplot.ylabel('y', fontsize=16)
pyplot.xlim(x_start/2, x_end/2)
pyplot.ylim(y_start/2, y_end/2)
ax2.quiver(xz3[::3,::3], yz3[::3,::3], uw[::3,::3], vw[::3,::3], units = 'inches', scale=6.0)
ax1.quiver(XRT[::3,::3], YRT[::3,::3], u[::3,::3], v[::3,::3], units = 'inches', scale=6.0)

# C_p
fig, (ax1, ax2) = pyplot.subplots(1, 2, figsize=(20, 8), sharex=True, sharey=True)
pyplot.suptitle('C_p for Part 4', size=24)
pyplot.xlabel('x', fontsize=16)
pyplot.ylabel('y', fontsize=16)
pyplot.xlim(1.25*x_start, 1.25*x_end)
pyplot.ylim(1.25*y_start, 1.25*y_end)
contf2 = ax2.contourf(xz3, yz3, cp, levels=numpy.linspace(-1,1,100), extend='both')
contf1 = ax1.contourf(XRT, YRT, cp1, levels=numpy.linspace(-1,1,100), extend='both')
fig.subplots_adjust(right=0.8)
cbar_ax = fig.add_axes([0.85, 0.15, 0.025, 0.7])
fig.colorbar(contf2, cax=cbar_ax, orientation='vertical')

print strength_doublet, uw[61,0], vw[61,0], numpy.min(cp[:,0])
```

```python
# Part 5
alpha = 20 * (2 * numpy.pi / 360)
origin = x0 + 1j * y0
ZRT = XRT + 1j * YRT
ZP = (ZRT - origin) * numpy.exp(-1j * alpha)
XP, YP = ZP.real, ZP.imag
x0p, y0p = 0., 0.

# doublet velocities
up, vp, psip = aero.get_doublet_info(strength_doublet, x0p, y0p, XP, YP)

# freestream
u_fs, v_fs, psi_fs = aero.get_freestream_info(u_inf, 0.0, XP, YP, Ntheta, Nr)

# combine
upc = (u_fs + up)
vpc = (v_fs + vp)
psi = psi_fs + psip

UP = numpy.empty_like(upc, dtype=complex)
UP = upc - 1j * vpc
dEdz = 1 - (c/ZRT)**2
wEp = UP / dEdz

Uz = UP * numpy.exp(-1j * alpha)
UwEp = wEp * numpy.exp(-1j * alpha)

uz = Uz.real
vz = -Uz.imag

upw = UwEp.real
vpw = -UwEp.imag

cpc = 1.0 - (uz**2 + vz**2) / u_inf**2
cpa = 1.0 - (upw**2 + vpw**2) / u_inf**2

vel_surf = numpy.sqrt(uz[:,0]**2 + vz[:,0]**2)
stags = numpy.where(numpy.abs(numpy.round(vel_surf,2)) == 0.0)
print stags, upw[49,0], vpw[49,0], cpa[74,0]

# Streamlines
fig, (ax1, ax2) = pyplot.subplots(1, 2, figsize=(16, 8), sharex=True, sharey=True)
pyplot.suptitle('Streamlines for Part 5', size=24)
pyplot.xlabel('x', fontsize=16)
pyplot.ylabel('y', fontsize=16)
pyplot.xlim(1.25*x_start, 1.25*x_end)
pyplot.ylim(1.25*y_start, 1.25*y_end)
ax2.contour(xz3, yz3, psi, colors='k', linestyles='solid', levels=numpy.linspace(numpy.min(psi),numpy.max(psi), 100), extend='True')
ax1.contour(XRT, YRT, psi, colors='k', linestyles='solid', levels=numpy.linspace(numpy.min(psi),numpy.max(psi), 100), extend='True')

# Velocity Vectors
fig, (ax1, ax2) = pyplot.subplots(1, 2, figsize=(16, 8), sharex=True, sharey=True)
pyplot.suptitle('Velocity Vectors for Part 5', size=24)
pyplot.xlabel('x', fontsize=16)
pyplot.ylabel('y', fontsize=16)
pyplot.xlim(x_start/2, x_end/2)
pyplot.ylim(y_start/2, y_end/2)
ax2.quiver(xz3[::3,::3], yz3[::3,::3], upw[::3,::3], vpw[::3,::3], units = 'inches', scale=6.0)
ax1.quiver(XRT[::3,::3], YRT[::3,::3], uz[::3,::3], vz[::3,::3], units = 'inches', scale=6.0)

# C_p
fig, (ax1, ax2) = pyplot.subplots(1, 2, figsize=(20, 8), sharex=True, sharey=True)
pyplot.suptitle('C_p for Part 5', size=24)
pyplot.xlabel('x', fontsize=16)
pyplot.ylabel('y', fontsize=16)
pyplot.xlim(1.25*x_start, 1.25*x_end)
pyplot.ylim(1.25*y_start, 1.25*y_end)
contf2 = ax2.contourf(xz3, yz3, cpa, levels=numpy.linspace(-1,1,100), extend='both')
contf1 = ax1.contourf(XRT, YRT, cpc, levels=numpy.linspace(-1,1,100), extend='both')
fig.subplots_adjust(right=0.8)
cbar_ax = fig.add_axes([0.85, 0.15, 0.025, 0.7])
fig.colorbar(contf2, cax=cbar_ax, orientation='vertical')
```

```python
# Part 6
alpha = 20 * (2 * numpy.pi / 360)
origin = x0 + 1j * y0
ZRT = XRT + 1j * YRT
ZP = (ZRT - origin) * numpy.exp(-1j * alpha)
XP, YP = ZP.real, ZP.imag
x0p, y0p = 0., 0.
strength_vortex = -numpy.sin(-alpha) * 4 * numpy.pi * r

# freestream
u_fs, v_fs, psi_fs = aero.get_freestream_info(u_inf, 0.0, XP, YP, Ntheta, Nr)

# doublet velocities
up, vp, psip = aero.get_doublet_info(strength_doublet, x0p, y0p, XP, YP)

# vortex velocities
u_vort, v_vort, psi_vort = aero.get_vortex_info(strength_vortex, x0p, y0p, XP, YP)

# combine
upc = (u_fs + up + u_vort)
vpc = (v_fs + vp + v_vort)
psi = psi_fs + psip + psi_vort

UP = numpy.empty_like(upc, dtype=complex)
UP = upc - 1j * vpc
dEdz = 1 - (c/ZP)**2
wEp = UP / dEdz

Uz = UP * numpy.exp(-1j * alpha)
UwEp = wEp * numpy.exp(-1j * alpha)

uz = Uz.real
vz = -Uz.imag

upw = UwEp.real
vpw = -UwEp.imag

cpc = 1.0 - (uz**2 + vz**2) / u_inf**2
cpa = 1.0 - (upw**2 + vpw**2) / u_inf**2

lift = - u_inf * strength_vortex * numpy.cos(alpha)
drag = (2 * numpy.pi * r / (Ntheta - 1)) * numpy.dot((0.5 * (1) * cpc[:,0] * u_inf**2), \
                                            numpy.cos(numpy.linspace(0,2*numpy.pi,Ntheta)))

vel_surf = numpy.sqrt(uz[:,0]**2 + vz[:,0]**2)
stags = numpy.where(numpy.abs(numpy.round(vel_surf,1)) == 0.0)

print strength_vortex, lift, drag, stags, upw[91,0], vpw[91,0], cpa[110,0]

# Streamlines
fig, (ax1, ax2) = pyplot.subplots(1, 2, figsize=(16, 8), sharex=True, sharey=True)
pyplot.suptitle('Streamlines for Part 6', size=24)
pyplot.xlabel('x', fontsize=16)
pyplot.ylabel('y', fontsize=16)
pyplot.xlim(1.25*x_start, 1.25*x_end)
pyplot.ylim(1.25*y_start, 1.25*y_end)
ax2.contour(xz3, yz3, psi, colors='k', linestyles='solid', levels=numpy.linspace(numpy.min(psi),numpy.max(psi), 100))
ax1.contour(XRT, YRT, psi, colors='k', linestyles='solid', levels=numpy.linspace(numpy.min(psi),numpy.max(psi), 100))

# Velocity Vectors
fig, (ax1, ax2) = pyplot.subplots(1, 2, figsize=(16, 8), sharex=True, sharey=True)
pyplot.suptitle('Velocity Vectors for Part 6', size=24)
pyplot.xlabel('x', fontsize=16)
pyplot.ylabel('y', fontsize=16)
pyplot.xlim(x_start/2, x_end/2)
pyplot.ylim(y_start/2, y_end/2)
ax2.quiver(xz3[::3,::3], yz3[::3,::3], upw[::3,::3], vpw[::3,::3], units = 'inches', scale=6.0)
ax1.quiver(XRT[::3,::3], YRT[::3,::3], uz[::3,::3], vz[::3,::3], units = 'inches', scale=6.0)

# C_p
fig, (ax1, ax2) = pyplot.subplots(1, 2, figsize=(20, 8), sharex=True, sharey=True)
pyplot.suptitle('C_p for Part 6', size=24)
pyplot.xlabel('x', fontsize=16)
pyplot.ylabel('y', fontsize=16)
pyplot.xlim(1.25*x_start, 1.25*x_end)
pyplot.ylim(1.25*y_start, 1.25*y_end)
contf2 = ax2.contourf(xz3, yz3, cpa, levels=numpy.linspace(-1,1,100), extend='both')
contf1 = ax1.contourf(XRT, YRT, cpc, levels=numpy.linspace(-1,1,100), extend='both')
fig.subplots_adjust(right=0.8)
cbar_ax = fig.add_axes([0.85, 0.15, 0.025, 0.7])
fig.colorbar(contf2, cax=cbar_ax, orientation='vertical')
```
