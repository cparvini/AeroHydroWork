```python
import math
import numpy
from matplotlib import pyplot
import aero_functions as aero
```

```python
N = 50
x_start, x_end = -2.0, 2.0
y_start, y_end = -1.0, 1.0
X, Y = aero.create_grid(N, x_start, x_end, y_start, y_end)
kappa = 1.0
gamma = 4.0
x_doublet, y_doublet = 0.0, 0.0
u_inf = 1.0
```

```python
# doublet velocities
u_doublet, v_doublet, psi_doublet = aero.get_doublet_info(kappa, x_doublet, y_doublet, X, Y)

# freestream
u_fs, v_fs, psi_fs = aero.get_freestream_info(u_inf, 0.0, X, Y, N)

# vortex
u_vortex, v_vortex, psi_vortex = aero.get_vortex_info(gamma, 0.0, 0.0, X, Y)

# combine
u = u_fs + u_doublet + u_vortex
v = v_fs + v_doublet + v_vortex
psi = psi_fs + psi_doublet + psi_vortex
```

```python
%matplotlib inline

# plots the streamlines
size = 10
pyplot.figure(figsize=(size, (y_end-y_start)/(x_end-x_start)*size))
pyplot.xlabel('x', fontsize=16)
pyplot.ylabel('y', fontsize=16)
pyplot.xlim(x_start, x_end)
pyplot.ylim(y_start, y_end)
pyplot.streamplot(X, Y, u, v, density=2, linewidth=1, arrowsize=1, arrowstyle='->')
pyplot.scatter(x_doublet, y_doublet, color='#CD2305', s=80, marker='o')

# calculates the cylinder radius and adds it to the figure
R = numpy.sqrt(kappa/(2*math.pi*u_inf))
circle = pyplot.Circle((0, 0), radius=R, color='#CD2305', alpha=0.5)
pyplot.gca().add_patch(circle)

# calculates the stagnation points and adds it to the figure
x_stagn1, y_stagn1 = +math.sqrt(R**2-(gamma/(4*math.pi*u_inf))**2), -gamma/(4*math.pi*u_inf)
x_stagn2, y_stagn2 = -math.sqrt(R**2-(gamma/(4*math.pi*u_inf))**2), -gamma/(4*math.pi*u_inf)

# limiting case: at values greater than 4*pi*u_inf*R, the stagnation points meet at the bottom
gammaLim = 4 * numpy.pi * u_inf * R
print(u'For values at or above \u0393 = %.3f, the Stagnation Points meet at the bottom then move off of the cylinder'\
      %(gammaLim))

th_stag = numpy.arcsin(-gamma/(2 * numpy.pi) / (2 * u_inf * R))
x_stag1a, y_stag1a = +numpy.cos(th_stag) * R, numpy.sin(th_stag) * R
x_stag2a, y_stag2a = -numpy.cos(th_stag) * R, numpy.sin(th_stag) * R


pyplot.scatter([x_stag1a, x_stag2a], [y_stag1a, y_stag2a], color='g', s=80, marker='o');
```

```python
# vtheta on the cylinder surface
theta = numpy.linspace(0, 2*math.pi, 100)
u_theta = -2*u_inf*numpy.sin(theta) - gamma/(2*math.pi*R)

# find cp
cp = 1.0 - (u_theta/u_inf)**2

# no vortex
u_theta_no_vortex = -2*u_inf*numpy.sin(theta)
cp_no_vortex = 1.0 - (u_theta_no_vortex/u_inf)**2

# cp plot
size = 6
pyplot.figure(figsize=(size, size))
pyplot.grid(True)
pyplot.xlabel('$\theta$', fontsize=18)
pyplot.ylabel('$C_p$', fontsize=18)
pyplot.xlim(theta.min(), theta.max())
pyplot.plot(theta, cp, color='#CD2305', linewidth=2, linestyle='-')
pyplot.plot(theta, cp_no_vortex, color='g', linewidth=2, linestyle='-')
pyplot.legend(['with vortex', 'without vortex'], loc='best', prop={'size':16});
```
