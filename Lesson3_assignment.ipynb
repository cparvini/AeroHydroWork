```python
import numpy
import math
import matplotlib.pyplot as plt
%matplotlib inline
from pylab import rcParams
rcParams['figure.figsize'] = [20.0, 16.0]
rcParams['font.size'] = 16
```

```python
def get_velocity_source(strength, xs, ys, X, Y):
    """Returns the velocity field generated by a source/sink.
    
    Arguments
    ---------
    strength -- strength of the source/sink.
    xs, ys -- coordinates of the source/sink.
    X, Y -- mesh grid.
    """
    
    u = strength/(2*math.pi)*(X-xs)/((X-xs)**2+(Y-ys)**2)
    v = strength/(2*math.pi)*(Y-ys)/((X-xs)**2+(Y-ys)**2)
    
    return u, v
```

```python
def get_freestream_info(u_inf, Y, N):
    """Returns the stream-function generated by a freestream.
    
    Arguments
    ---------
    u_inf -- freestream velocity.
    Y -- mesh grid of y values.
    N -- number of discrete points in each direction on our grid.
    """
    u = u_inf * numpy.ones((N, N), dtype=float)
    v = numpy.zeros((N, N), dtype=float)
    psi = u_inf * Y
    
    return u, v, psi
```

```python
def get_stream_function_source(strength, xs, ys, X, Y):
    """Returns the strem-function generated by a source/sink.
    
    Arguments
    ---------
    strength -- strength of the source/sink.
    xs, ys -- coordinates of the source/sink.
    X, Y -- mesh grid.
    """
    psi = strength/(2*pi)*numpy.arctan2((Y-ys), (X-xs))
    
    return psi
```

```python
def combine_sources(sigma, x_source, y_source, X, Y, N):
    """Returns the combined velocities and stream-function of a series of
    sources or sinks.
    
    Arguments
    ---------
    sigma -- float; strength of the source/sink.
    xs, ys -- 1D array of flot; coordinates of the sources/sinks.
    X, Y -- 2D matrix of float; mesh grid.
    N -- number of grid points.
    
    Returns
    ---------
    u_source_all, v_source_all, psi_source_all -- 2D matrix of float
    """
    
    u_source_all, v_source_all = numpy.zeros((N,N)), numpy.zeros((N,N))
    psi_source_all = numpy.zeros((N,N))
    
    for i in range(0,len(sigma)):
    
        u_source, v_source = get_velocity_source\
                            (sigma[i],x_source[i],y_source[i],X,Y)
        psi_source = get_stream_function_source\
                        (sigma[i],x_source[i],y_source[i],X,Y)
    
        u_source_all += u_source
        v_source_all += v_source
        psi_source_all += psi_source
    
    return u_source_all, v_source_all, psi_source_all
```

```python
# loading
x_in = numpy.loadtxt('NACA0012_x.txt')
y_in = numpy.loadtxt('NACA0012_y.txt')
sigma_in = numpy.loadtxt('NACA0012_sigma.txt')
```

```python
N = 51
x_start, x_end = -1.0, 2.0
y_start, y_end = -0.5, 0.5
x = numpy.linspace(x_start, x_end, N)
y = numpy.linspace(y_start, y_end, N)
X, Y = numpy.meshgrid(x,y)

u_fs, v_fs, psi_fs = numpy.zeros((N,N)), numpy.zeros((N,N)), numpy.zeros((N,N))
u_ss, v_ss, psi_ss = numpy.zeros((N,N)), numpy.zeros((N,N)), numpy.zeros((N,N))
u_inf = 1.0
```

```python
# the math
u_fs, v_fs, psi_fs = get_freestream_info(u_inf, Y, N)
u_ss, v_ss, psi_ss = combine_sources(sigma_in, x_in, y_in, X, Y, N)

u = u_fs + u_ss
v = v_fs + v_ss
psi = psi_fs + psi_ss

cp = 1.0 - (u**2 + v**2)/u_inf**2

cp_max = numpy.max(cp)
max_loc = numpy.argmax(cp)
print('The maximum C_p value is %.4f and located at x = %.2f, y = %.2f' \
      %(cp_max,X.flat[max_loc],Y.flat[max_loc]))

size = 10
pyplot.xlabel('x', fontsize=16)
pyplot.ylabel('y', fontsize=16)
pyplot.xlim(x_start, x_end)
pyplot.ylim(y_start, y_end)
contf = pyplot.contourf(X,Y,cp,levels=numpy.linspace(-0.6,0.5,20))
pyplot.colorbar(contf, orientation='horizontal')
pyplot.streamplot(X, Y, u, v, density=2.0, linewidth=1, arrowsize=1, arrowstyle='->')
pyplot.scatter(x_in, y_in, color='k', s=40, marker='o');
pyplot.scatter(X.flat[max_loc],Y.flat[max_loc], color='w', s=300, \
               marker='x', linewidth=5.0)
```
