```python
import numpy
import math
import aero_functions as aero
from aero_functions import Source
import scipy
from matplotlib import pyplot
%matplotlib inline
```

```python
def get_sourceline_velocity(strength_source, x_source, y_source, N_sources, X, Y):
    """Returns the cartesian velocities generated by a source line.
    
    Arguments
    ---------
    strength_source -- float; fraction of the total source strength.
    x_source, y_source -- 1D Array of float; cartesian source locations.
    N_sources -- int; number of sources in the line.
    X, Y -- 2D matrix of float; mesh grid of x and y values.
    
    Output
    ---------
    sources -- source class; source class object.
    """
    sources = numpy.empty(N_sources, dtype=object)

    for i in xrange(N_sources):
        sources[i] = Source(strength_source, x_source[i], y_source[i])
        sources[i].velocity(X, Y)
        sources[i].stream_function(X, Y)
    
    return sources
```

```python
# grid
N = 100
xi, xf, yi, yf = -1.0, 1.0, -1.5, 1.5
X, Y = aero.create_grid(N, xi, xf, yi, yf)

# variables
u_inf = 1.0
```

```python
# flow variables
N_sources = 11
strength = 5.
strength_source = strength/N_sources
x_source = numpy.zeros(N_sources, dtype=float)
y_source = numpy.linspace(-1.0, 1.0, N_sources)

# flow calculations
u_fs, v_fs, psi_fs = aero.get_freestream_info(u_inf, 0.0, X, Y, N)
sources = get_sourceline_velocity(strength_source, x_source, y_source, N_sources, X, Y)
u = u_fs.copy()
v = v_fs.copy()
psi = psi_fs.copy()

for source in sources:
    u += source.u
    v += source.v
    psi += source.psi
    
# plots the streamlines
%matplotlib inline

size = 6
pyplot.figure(figsize=(size, (yf-yi)/(xf-xi)*size))
pyplot.grid(True)
pyplot.xlabel('x', fontsize=16)
pyplot.ylabel('y', fontsize=16)
pyplot.xlim(xi, xf)
pyplot.ylim(yi, yf)
pyplot.streamplot(X, Y, u, v, density=2, linewidth=1, arrowsize=1, arrowstyle='->')
pyplot.scatter(x_source, y_source, color='#CD2305', s=80, marker='o')
#ids = numpy.ravel_multi_index(numpy.where(numpy.round(psi,1) == 0), psi.shape)
#pyplot.scatter(X.flat[ids], Y.flat[ids], color='#CD2305', s=80, marker='o')
pyplot.contour(X, Y, psi, levels=[0.], colors='#CD2305', linewidths=2, linestyles='solid')
velocity = pyplot.contourf(X, Y, numpy.sqrt(u**2+v**2), levels=numpy.linspace(0.0, 0.1, 10))
cbar = pyplot.colorbar(velocity, ticks=[0, 0.05, 0.1], orientation='horizontal')
cbar.set_label('Velocity magnitude', fontsize=16);
```

```python
sigma = 2 * u_inf
y_min, y_max = -1.0, 1.0
u_sheet = numpy.empty((N,N), dtype=float)
v_sheet = numpy.empty((N,N), dtype=float)
psi_sheet = numpy.empty((N,N), dtype=float)
for i in xrange(N):
    for j in xrange(N):
        integrand = lambda s : X[i,j]/(X[i,j]**2 + (Y[i,j] - s)**2)
        u_sheet[i,j] = sigma/(2*math.pi)*integrate.quad(integrand, y_min, y_max)[0]
        
        integrand = lambda s : (Y[i,j] - s)/(X[i,j]**2 + (Y[i,j] - s)**2)
        v_sheet[i,j] = sigma/(2*math.pi)*integrate.quad(integrand, y_min, y_max)[0]
        
        integrand = lambda s : scipy.arctan2((Y[i,j] - s),X[i,j])
        psi_sheet[i,j] = sigma/(2*math.pi)*integrate.quad(integrand, y_min, y_max)[0]
        
u = u_fs + u_sheet
v = v_fs + v_sheet
psi = psi_fs + psi_sheet

# plots the streamlines
%matplotlib inline

size = 6
pyplot.figure(figsize=(size, (yf-yi)/(xf-xi)*size))
pyplot.grid(True)
pyplot.xlabel('x', fontsize=16)
pyplot.ylabel('y', fontsize=16)
pyplot.xlim(xi, xf)
pyplot.ylim(yi, yf)
pyplot.streamplot(X, Y, u, v, density=2, linewidth=1, arrowsize=1, arrowstyle='->')
pyplot.axvline(0.0, (y_min-yi)/(yf-yi), (y_max-yi)/(yf-yi), 
            color='#CD2305', linewidth=4)
pyplot.contour(X, Y, psi, levels=[0.], colors='#CD2305', linewidths=2, linestyles='solid')
#ids = numpy.ravel_multi_index(numpy.where(numpy.round(psi,1) == 0), psi.shape)
#pyplot.scatter(X.flat[ids], Y.flat[ids], color='#CD2305', s=80, marker='o')
velocity = pyplot.contourf(X, Y, numpy.sqrt(u**2+v**2), levels=numpy.linspace(0.0, 0.1, 10))
cbar = pyplot.colorbar(velocity, ticks=[0, 0.05, 0.1], orientation='horizontal')
cbar.set_label('Velocity magnitude', fontsize=16);
```

The stagnation point is now on the surface of the source sheet when: 

$$\sigma = 2 u_{\infty}$$

The dividing streamline still appears at y = 0. This is starting to make sense, because the streamfunction for a freestream is simply:

$$\psi_{freestream} = u_{\infty} y$$

This means that the 0 potential line for a freestream \textit{must} be on the y = 0 line, and therefore the addition of the two stream functions is leaving the y = 0 line unchanged. Is there a problem with the stream function additions from the source lines?
