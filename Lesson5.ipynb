```python
import numpy
import math
from matplotlib import pyplot
%matplotlib inline
pyplot.rcParams['font.size'] = 16
```

```python
def get_velocity_vortex(strength, xv, yv, X, Y):
    """Returns the velocity field generated by a vortex.
    
    Arguments
    ---------
    strength -- float; strength of the vortex.
    xv, yv -- float; coordinates of the vortex.
    X, Y -- 2D matrix; mesh grid.
    
    Output
    ---------
    u, v -- 2D matrix; cartesian velocities of the vortex.
    """
    u = + strength/(2*math.pi)*(Y-yv)/((X-xv)**2+(Y-yv)**2)
    v = - strength/(2*math.pi)*(X-xv)/((X-xv)**2+(Y-yv)**2)
    
    return u, v
```

```python
def get_stream_function_vortex(strength, xv, yv, X, Y):
    """Returns the stream-function generated by a vortex.
    
    Arguments
    ---------
    strength -- float; strength of the vortex.
    xv, yv -- float; coordinates of the vortex.
    X, Y -- 2D array; mesh grid.
    
    Output
    ---------
    psi -- 2D matrix; potential values on a grid.
    """
    psi = strength/(2*math.pi)*numpy.log(numpy.sqrt((X-xv)**2+(Y-yv)**2))
    
    return psi
```

```python
def combine_vortices(gamma, x_vortices, y_vortices, X, Y, N):
    """Returns the combined velocities and stream-function of a series of vortices.
    
    Arguments
    ---------
    gamma -- strength of the vortex.
    xv, yv -- coordinates of the vortex.
    X, Y -- mesh grid.
    
    Output
    ---------
    u_vortex_row,
    v_vortex_row,
    psi_vortex_row -- 2D matrix; cartesian velocities and stream function
                      on a grid.
    """
    
    u_vortex, v_vortex = numpy.zeros((N,N)), numpy.zeros((N,N))
    u_vortex_row, v_vortex_row = numpy.zeros((N,N)), numpy.zeros((N,N))
    psi_vortex_row = numpy.zeros((N,N))
    
    for i  in range(0,n-1):
    
        u_vortex, v_vortex = get_velocity_vortex(gamma,x_vortices[i],y_vortices[i],X,Y)
        psi_vortex = get_stream_function_vortex(gamma,x_vortices[i],y_vortices[i],X,Y)
    
        u_vortex_row += u_vortex
        v_vortex_row += v_vortex
        psi_vortex_row += psi_vortex
    
    return u_vortex_row, v_vortex_row, psi_vortex_row
```

```python
def get_freestream_info(u_inf, Y, N):
    """Returns the stream-function generated by a freestream.
    
    Arguments
    ---------
    u_inf -- float; freestream velocity.
    Y -- 2D matrix; mesh grid of y values.
    N -- int; number of discrete points in each direction on our grid.
    
    Output
    ---------
    u, v -- 2D matrix; cartesian velocities of the vortex.
    psi -- 2D matrix; stream-function on a grid.
    """
    u = u_inf * numpy.ones((N, N), dtype=float)
    v = numpy.zeros((N, N), dtype=float)
    psi = u_inf * Y
    
    return u, v, psi
```

```python
def get_velocity_infinite_vortices(gamma,X,Y,x_vortices):
    """Returns the cartesian velocities of an infinite row of vortices.
    
    Arguments
    ---------
    gamma -- float; freestream velocity.
    Y -- 2D matrix; mesh grid of y values.
    N -- int; number of discrete points in each direction on our grid.
    
    Output
    ---------
    u, v -- 2D matrix; cartesian velocities of the vortex.
    """    
    
    a = x_vortices[1]-x_vortices[0]
    c = 2*math.pi/a
    
    u = + (gamma/(2*a))*((numpy.sinh(c*Y))/(numpy.cosh(c*Y) - numpy.cos(c*X)))
    v = - (gamma/(2*a))*((numpy.sin(c*X))/(numpy.cosh(c*Y) - numpy.cos(c*X)))
    
    return u, v
```

```python
def create_grid(N,x_start,x_end,y_start,y_end):
    """Creates a 2D meshgrid based on input values.
    
    Arguments
    ---------
    N -- int; number of gridpoints.
    x_start,x_end -- float; x-direction boundaries for the grid.
    y_start,y_end -- float; y-direction boundaries for the grid.
    
    Output
    ---------
    X, Y -- 2D matrix; meshgrids of x and y coordinates.
    """
    
    x = numpy.linspace(x_start, x_end, N)
    y = numpy.linspace(y_start, y_end, N)
    X, Y = numpy.meshgrid(x,y)
    
    return X, Y
```

```python
# initial setup
N = 100
n = 24
x_start, x_end = -2.0, 2.0
y_start, y_end = -0.5, 0.5
x = numpy.linspace(x_start, x_end, N)
y = numpy.linspace(y_start, y_end, N)
X, Y = numpy.meshgrid(x,y)

u_vortex, v_vortex = numpy.zeros((N,N)), numpy.zeros((N,N))
u_vortex_row, v_vortex_row = numpy.zeros((N,N)), numpy.zeros((N,N))
psi_vortex_row = numpy.zeros((N,N))
gamma = 1.0
u_inf = 4.0
x_vortices = x[::N/(n-1)]
y_vortices = numpy.zeros_like(x_vortices)
```

```python
# the math
u, v, psi = combine_vortices(gamma, x_vortices, y_vortices, X, Y, N)
u_fs, v_fs, psi_fs = get_freestream_info(u_inf, Y, N)
# u += u_fs
# v += v_fs
# psi += psi_fs
u_all, v_all = get_velocity_infinite_vortices(gamma, X, Y, x_vortices)
```
